# **EXTRAÇÃO**

```python
import pandas as pd
import numpy as np
import requests
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/refs/heads/main/TelecomX_Data.json'
```

```python
response = requests.get(url)
```

```python
df = pd.DataFrame(response.json())
df.head()
```

```python
df.to_csv("data_df.csv", index=False)
```

```python
df_normalizado = pd.json_normalize(df.to_dict(orient='records'), sep='_')
df_normalizado.head()
```

# **IMFORMAÇÃO**

```python

def remove_prefix_string(text):
  """
  Remove o prefixo antes do primeiro '_' em uma string usando manipulacao de string.

  Args:
    text: A string de entrada.

  Returns:
    A string sem o prefixo e o '_', ou a string original se não houver '_'.
  """
  # Verifica se a entrada é uma string antes de aplicar a logica
  if isinstance(text, str):
    if '_' in text:
      return text.split('_', 1)[1]
  # Retorna a entrada original se não for uma string ou nao contiver '_'
  return text

# Renomear as colunas do DataFrame df_normalizado
df_normalizado = df_normalizado.rename(columns=lambda x: remove_prefix_string(x))

# Para visualizar as colunas renomeadas
print("Colunas do DataFrame após renomear:")
print("-" * 30)
for i, col in enumerate(df_normalizado.columns):
  print(f"{i}. {col}")
print("-" * 30)

# Para visualizar as primeiras linhas do DataFrame atualizado
df_normalizado.head()
```

```python
print(f'Linhas: {df_normalizado.shape[0]}\nColunas: {df_normalizado.shape[1]}')
```

```python
df_normalizado.info()
```

```python
for col in df_normalizado.columns:
  print('-'*75)
  unique_values = df_normalizado[col].unique()
  if len(unique_values) > 10:  # Adjust this number to control how many unique values are displayed
    print(f'Coluna - {col}\nValores únicos - {unique_values[:10]}...')
  else:
    print(f'Coluna - {col}\nValores únicos - {unique_values}')
```

# **RELATÓRIO**


# 1. *Contexto*

Este relatório fornece uma análise detalhada do comportamento dos clientes da TelecomX, visando entender as razões que levam ao cancelamento de serviços (churn). O objetivo é fornecer insights estratégicos que permitam aumentar a retenção de clientes e reduzir a evasão.

# 2. *Estratégia Metodológica*

**Coleta de Informações:**
Os dados foram obtidos de arquivos no formato JSON.

**Preparação e Transformação:**

- Transformação do JSON em um DataFrame do pandas.

- Remoção de registros sem dados de churn.

- Conversão da coluna TotalCharges em valores numéricos, substituindo dados ausentes por 0.

- Criação da métrica diária de cobrança: Daily_Charge = MonthlyCharges / 30,42.

- Conversão de respostas "Yes/No" em valores binários (1/0) nas colunas de serviços e cobrança digital.

**Análise e Exploração:**

- Contagem de categorias e estatísticas descritivas.

- Visualizações com histogramas, boxplots e gráficos de barras.

- Cálculo de taxas de churn por segmento e análise de correlações entre variáveis.

# 3. *Principais Fatores de Churn*

**Os elementos mais significativos para a rescisão de serviços foram:**

- Clientes recentes (< 6 meses): maior probabilidade de cancelamento.

- Contratos mensais: taxa de churn de 42,7%, superior a contratos anuais (11,1%) e bienais (2,9%).

- Daily_Charge elevado: risco maior de evasão.

- Pagamento via Electronic Check: associado a maior número de cancelamentos.

- Baixo uso de serviços: clientes que não usam internet, suporte técnico ou backup têm maior probabilidade de churn.

- Insatisfação com serviços essenciais: problemas em OnlineSecurity e DeviceProtection aumentam o risco.

- Demografia: jovens/adultos sem parceiros ou dependentes apresentam taxas mais altas.

- Falta de Internet: ausência do serviço aumenta risco.

- Serviços contratados, mas não valorizados: DeviceProtection e TechSupport podem gerar churn devido à percepção de baixo valor.

- Custos elevados: tarifas mensais ou diárias altas impactam negativamente a percepção de custo-benefício.

#4.  *Sugestões Táticas*

**Integração e Monitoramento Inicial (0–6 meses)**

- Apoio próximo com verificações periódicas.

- Tutoriais breves e FAQ interativo para familiarização com os serviços.

**Estímulos para Acordos de Longa Duração**

- Desconto de 15% na renovação anual.

- Três meses gratuitos em contratos de dois anos.

**Pacotes de Atendimento**

- Pacote "Segurança Total": OnlineSecurity + TechSupport + Backup com 20% de desconto.

- Ressaltar claramente os benefícios dos pacotes.

**Aprimoramento no pagamento por meio de cheque eletrônico**

- Auditar e otimizar o processo de cobrança.

- Incentivar migração para débito automático com 5% de desconto.

**Campanhas direcionadas**

- Foco em novos clientes, jovens/adultos e usuários de fatura digital.

- Enviar ofertas personalizadas antes do término do contrato.

**Manutenção de Clientes de Alto Risco**

- Ações voltadas para jovens sem parceiros/dependentes, clientes que usam Electronic Check e contratos mensais.

- Ofertas adaptáveis, suporte proativo e pacotes que reforcem o valor percebido.

**Melhoria da Qualidade dos Serviços**

- Pesquisas de satisfação e auditorias em OnlineSecurity, DeviceProtection e TechSupport.

- Campanhas informativas sobre o valor agregado desses serviços.

**Ofertas de Pagamento Estável**

- Estimular pagamentos por meio de cartão ou débito automático

- Ressaltar a segurança e a praticidade do método.

**Harmonia entre Custo e Valor**

- Analisar clientes que pagam mais do que percebem como benefício.

- Oferecer pacotes personalizados ou destacar o retorno sobre o investimento.

#5. *Conclusão e Etapas Futuras*

A análise indicou que o churn está relacionado a tipo de contrato, forma de pagamento, tempo de permanência e uso de serviços. Além disso, fatores demográficos e a percepção de valor dos serviços contratados influenciam significativamente a decisão de cancelamento.

**Próximas ações recomendadas:**

- Desenvolver modelo preditivo de churn para identificar clientes em risco.

- Implementar estratégias segmentadas com base nos fatores críticos.

- Monitorar trimestralmente a taxa de churn e ajustar ações com base em KPIs e feedback dos clientes.
